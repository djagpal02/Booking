```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import math
%matplotlib inline
```

```python
df = pd.read_csv('booking_train_set.csv')
```

```python
df.drop(['index'],axis=1,inplace = True);
```

```python
# Do people using particular devices book more to particular cities?
```

```python
# Get Labels for Cities

df["city_id"] = df["city_id"].astype('category')
df["cityLabel"] = df["city_id"].cat.codes
```

```python
# Get Labels for Deices

df["device_class"] = df["device_class"].astype('category')
df["deviceLabel"] = df["device_class"].cat.codes
```

```python
df_new = df[['deviceLabel', 'cityLabel']]
```

```python
sns.jointplot(y='cityLabel', x='deviceLabel', data=df_new, kind='hex')
```

```python
dev_city = pd.crosstab(df.deviceLabel,df.cityLabel)
dev_city.head()
```
